```python
import requests
response_object=requests.get('https://emlii.com/78-events-across-100-years-that-completely-changed-the-world/')
response_object.text 
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(response_object.text,'html.parser')
soup
```

```python
'''TASK1-EXTRACTING SUBTITLE'''
events=soup.find_all('h2',attrs={'class':'article-subtitle'})
events
```
